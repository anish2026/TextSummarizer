```python
from transformers import pipeline
#there are many ways of fetching dataset and one of the most useful way is to do via pipeline
```

```python
import pandas as pd
from simpletransformers.seq2seq import Seq2SeqModel,Seq2SeqArgs

#we are using Seq2SeqModel
#summarizers requires seq2seq module and further we will make functions for that
```

```python
model_args = Seq2SeqArgs() #it is nothng but the hyper parameter we will use for our model
model_args.num_train_epochs = 10 #epoch value can be taken whatever we want because this is the responsible for the accuracy
model_args.no_save = True
model_args.evaluate_generated_text = True
model_args.evaluate_during_training = True
model_args.evaluate_during_training_verbose = True

# I am just setting all values to True by default


model = Seq2SeqModel(
    encoder_decoder_type="bart",
    encoder_decoder_name="facebook/bart-large",
    args=model_args,
    use_cuda=False,
)

#For text summarization , we need both encoder and decoder transformers
# Bart is such model
# for now we are stick with english language.


```

```python
train_data = [
    [
        "Anish Anand is an Indian.He is ML enthusiastic person.He is persuing his Btech from Vellore Institute of technology.",
        "Anish Anand is a btech student and Ml enthusiastic.",
    ],
    [
        "Anish has two brothers.His elder brother is a computer science engineer in Google.The younger brother of Anish is actually a twin brother of him,he is also persuing Btech from Jaipur college.",
        "Anish has two brothers and both of them are engineer in google and Jaipur respectively.",
    ],
]


#forming my dataset in data frame
train_df = pd.DataFrame(
    train_data, columns=["input_text", "target_text"]
)

# I am not using huge dataset as for now, just making my own dataset.
# The input text is the whole text and the target is nothing but summary of it.
```

```python
eval_data = [
    [
        "Anish is very nice boy , friendly to all and always help everyone therefore he is the president of many clubs in his school.",
        "Anish is a good leader and helping person.",
    ],
    [
        "Professors say about Anish that he is an extraordinary guy with different art of vision and he will success in his life soon.",
        "Someday Anish will become a great man who has a vision for society.",
    ],
]

eval_df = pd.DataFrame(eval_data, columns=["input_text", "target_text"])

#stored them in a new dataset after testing 
```

```python
# Train the model
model.train_model(train_df, eval_data=eval_df)
```
